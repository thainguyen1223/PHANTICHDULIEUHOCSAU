```python
import numpy as numpy
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/quangvinht/Phan_tich_du_lieu_hoc_sau/main/dulieuxettuyendaihoc.csv')
print(df)

```

```python
 print("in ra 10 dòng đầu:") 
 print(df.head(10))
```

```python
print("in ra 10 dòng cuối:") 
print(df.tail(10))
```

```python
#Câu 4:
df['DT'].fillna( 0 , inplace= True )
#Câu 5: 
df['T1'].fillna( df['T1'].mean() , inplace= True )
print(df)
```

```python
#cau 6:

def fill(col):
  return df[col].fillna( df[col].mean() , inplace= True  )

isfloat64=df.dtypes == float
df_column_list = df.columns[isfloat64]

for col in df_column_list:
  fill(col)

print(df.head(10))


```

```python
#câu 7:
for i in ['1','2','6']:
  df['TBM'+i] = (df['T'+i]*2 + df['L'+i]+ df['H'+i]+ df['S'+i]+ df['V'+i]*2 + df['X'+i]+ df['D'+i]+ df['N'+i]) / 10

df.rename(columns={'TBM6':'TBM3'})
```

```python
# for i in ['1','2','3']:
  # if (df['TBM'+i] < 5):
  #     df['XL'+i] = 'Y'
  # if df['TBM'+i] >= 5 and df['TBM'+i] < 6.5:
  #     df['XL'+i] = 'TB'
  # if df['TBM'+i] >= 6.5 and df['TBM'+i] < 8: 
  #     df['XL'+i] = 'K'
  # if df['TBM'+i] >= 8 and df['TBM'+i] < 9: 
  #     df['XL'+i] = 'G'
  # if df['TBM'+i] > 9: 
  #     df['XL'+i] = 'XS'

#câu 8:
df.loc[ (df['TBM1']  < 5) , 'XL1' ] = 'Y'
df.loc[ (df['TBM1'] >= 5) & (df['TBM1'] < 6.5) , 'XL1'] = 'TB'
df.loc[ (df['TBM1'] >= 6.5) & (df['TBM1'] < 8) , 'XL1'] = 'K'
df.loc[ (df['TBM1'] >= 8) & (df['TBM1'] < 9) , 'XL1'] = 'G'
df.loc[(df['TBM1']  > 9) , 'XL1'] = 'XS'

df.loc[ (df['TBM2']  < 5) , 'XL2' ] = 'Y'
df.loc[ (df['TBM2'] >= 5) & (df['TBM2'] < 6.5) , 'XL2'] = 'TB'
df.loc[ (df['TBM2'] >= 6.5) & (df['TBM2'] < 8) , 'XL2'] = 'K'
df.loc[ (df['TBM2'] >= 8) & (df['TBM2'] < 9) , 'XL2'] = 'G'
df.loc[(df['TBM2']  > 9) , 'XL2'] = 'XS'


df.loc[ (df['TBM6']  < 5) , 'XL3' ] = 'Y'
df.loc[ (df['TBM6'] >= 5) & (df['TBM6'] < 6.5) , 'XL3'] = 'TB'
df.loc[ (df['TBM6'] >= 6.5) & (df['TBM6'] < 8) , 'XL3'] = 'K'
df.loc[ (df['TBM6'] >= 8) & (df['TBM6'] < 9) , 'XL3'] = 'G'
df.loc[(df['TBM6']  > 9) , 'XL3'] = 'XS'

print(df)
```

```python
#câu 9 :
for i in ['1','2','6']:
  df['US_TMB'+i] = (df['TBM'+i]/10) * 4
print(df['US_TMB1'])
 

```
