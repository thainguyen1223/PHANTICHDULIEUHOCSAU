```python
datatitanic = pd.read_csv("https://raw.githubusercontent.com/huynhhoc/phantichdulieu/main/Data/titanic_disaster.csv")
datatitanic.head(10)
```

```python
import seaborn as sns

sns.heatmap(datatitanic.isna(), yticklabels=False, cbar=True, cmap='viridis')
```

```python
datatitanic[['firstName ', 'secondName']]= datatitanic['Name'].str.split(pat=',',expand=True)


datatitanic = datatitanic.drop('Name',axis=1)
```

```python
datatitanic.head(10)
```

```python
gender = {'male': 'M', 'female': 'F'}
datatitanic['Sex'].replace(gender,inplace=True)
datatitanic.head(10)
```

```python
sns.heatmap(datatitanic.isna(), 
            yticklabels=False, 
            cbar=True, 
            cmap='viridis')
```

```python
index_agenull = datatitanic[datatitanic['Age'].isna()].index
index_agenull
```

```python
datatitanic.drop(index_agenull, inplace=True)
```

```python
sns.heatmap(datatitanic.isna(), 
            yticklabels=False, 
            cbar=True, 
            cmap='viridis')
```

```python
datatitanic["Age"][datatitanic["Pclass"] == 1] = datatitanic["Age"][datatitanic["Pclass"] == 1].fillna(datatitanic["Age"][datatitanic["Pclass"] == 1].median(), inplace = False)
```

```python
datatitanic.loc[ (datatitanic['Age']  < 12) , 'Agegroup ' ] = 'Kid'
datatitanic.loc[ (datatitanic['Age'] > 12) & (datatitanic['Age'] <= 18) , 'Agegroup '] = 'Teen'
datatitanic.loc[ (datatitanic['Age'] > 18) & (datatitanic['Age'] <= 60) , 'Agegroup '] = 'Adult'
datatitanic.loc[ (datatitanic['Age']  > 60) , 'Agegroup ' ] = 'Older'
```

```python
datatitanic.head(10)
```

```python


datatitanic.loc[ (datatitanic['secondName'].str.find('Mr')) == 1 , 'namePrefix ' ] = 'Mr'
datatitanic.loc[ (datatitanic['secondName'].str.find('Mrs')) == 1 , 'namePrefix ' ] = 'Mrs'
datatitanic.loc[ (datatitanic['secondName'].str.find('Miss')) == 1 , 'namePrefix ' ] = 'Miss'
datatitanic.loc[ (datatitanic['secondName'].str.find('Master')) == 1 , 'namePrefix ' ] = 'Master'

```

```python
indexSecondName=datatitanic.index
secondName = datatitanic['secondName']

for i in indexSecondName:
  s = secondName[i].split('.')
  secondName[i] = s[1]

datatitanic.head(10)

 

```

```python
datatitanic['familySize'] = 0

for i in indexSecondName:
  datatitanic['familySize'][i] = 1 + datatitanic['SibSp'][i] + datatitanic['Parch'][i]

datatitanic.head(10)
```

```python
datatitanic.loc[ (datatitanic['familySize'] == 0) , 'Alone ' ] = 1
datatitanic.loc[ (datatitanic['familySize'] > 0 ) , 'Alone ' ] = 0

datatitanic.head(10)
```

```python
datatitanic['Cabin'].fillna('Unknown', inplace =True)

datatitanic['typeCabin'] = datatitanic['Cabin']

for i in indexSecondName:
  if datatitanic['typeCabin'][i] != 'Unknown':
      datatitanic['typeCabin'][i] = datatitanic['typeCabin'][i][0]

datatitanic.head(10)



```
